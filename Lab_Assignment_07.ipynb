============================
Function Call:
function_name('HaRRy PotteR')
Output:

No. of Lowercase Characters: 6

```python
def num_of_up_low(string):
    upchr = ''
    lowchr = ''
    for i in string:
        if 65 <= ord(i) <= 90:
            upchr += i
        elif 97 <= ord(i) <= 122:
            lowchr += i
        else:
            pass
    print("No. of Uppercase characters:",len(upchr))
    print("No. of Lowercase characters:",len(lowchr))

string = input("Enter a string: ")
num_of_up_low(string)
```
